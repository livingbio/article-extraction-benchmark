```python
import json
```

```python
with open('./newoutput/glia.json', 'r') as in_file:
    data = json.load(in_file)
```

```python
def filter_content(text):
    t = text
    while t.find('![') != -1:
        start = t.find('![')
        step = 1
        stack = 0
        while t[start + step: start + step + 5] != '(http':
            step = step + 1
        stack = stack + 1


        while stack != 0:
            step += 1
            if t[start + step] == '(':
                stack = stack + 1
            elif t[start + step] == ')':
                stack = stack - 1 
        end = start + step
        t = t[:start] + t[end+1:]
    return t
```

```python
for key, value in data.items():
    data[key]['articleBody'] = filter_content(data[key]['articleBody'])
```

```python
with open('./newoutput/filter_glia.json', 'w') as out_file:
    json.dump(data, out_file, ensure_ascii=False, indent=4)
```
